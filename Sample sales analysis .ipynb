#### Step 2: Overall Sales and Profit Summary (KPIs)

```python
# Calculate KPIs
total_sales = superstore_data['Sales'].sum()
total_profit = superstore_data['Profit'].sum()
num_orders = superstore_data.shape[0]
average_order_value = total_sales / num_orders
gross_margin = (total_profit / total_sales) * 100
```

```python
# Print KPIs
print(f"Total Sales: {total_sales}")
print(f"Total Profit: {total_profit}")
print(f"Average Order Value (AOV): {average_order_value}")
print(f"Gross Margin: {gross_margin:.2f}%")
```

#### Step 3: Monthly Sales by District

###### monthly sales across different districts or regions

```python
# Extract Order Month and Region/District columns
superstore_data['Order Month'] = superstore_data['Order Date'].dt.to_period('M')
district_monthly_sales = superstore_data.groupby(['Region', 'Order Month'])['Sales'].sum().unstack().fillna(0)
```

```python
# Plot district monthly sales
plt.figure(figsize=(14, 8))
sns.heatmap(district_monthly_sales, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title('Monthly Sales by Region')
plt.xlabel('Month')
plt.ylabel('Region')
plt.show()
```

#### Step 4: New Store Analysis

###### the count of new stores added each year 

```python
# Extract year from opening date (assuming 'Store Opening Date' column exists)
superstore_data['Store Opening Year'] = superstore_data['Order Date'].dt.year  # Replace with actual column
```

```python
# Count of new stores per year
new_stores_per_year = superstore_data.groupby('Store Opening Year').size()
```

```python
plt.figure(figsize=(10, 5))
new_stores_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of New Stores Opened Each Year')
plt.xlabel('Year')
plt.ylabel('Number of New Stores')
plt.show()
```

#### Step 5: Regular Sales vs. Markdown Sales Analysis

###### sales between regular-priced items and markdowns.

```python
# Calculate regular and markdown sales
regular_sales = superstore_data[superstore_data['Discount'] == 0]['Sales'].sum()
markdown_sales = superstore_data[superstore_data['Discount'] > 0]['Sales'].sum()
```

```python
# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x=['Regular Sales', 'Markdown Sales'], y=[regular_sales, markdown_sales], palette='viridis')
plt.title('Regular vs. Markdown Sales')
plt.ylabel('Total Sales')
plt.show()
```

#### Step 6: This Year (TY) vs. Last Year (LY) Sales and Variance

```python
# Extract year and aggregate sales by year
superstore_data['Order Year'] = superstore_data['Order Date'].dt.year
yearly_sales = superstore_data.groupby('Order Year')['Sales'].sum()
```

```python
# Calculate TY, LY, and variance
ty_sales = yearly_sales.iloc[-1]
ly_sales = yearly_sales.iloc[-2] if len(yearly_sales) > 1 else 0
variance = ty_sales - ly_sales
```

```python
print(f"This Year (TY) Sales: {ty_sales}")
print(f"Last Year (LY) Sales: {ly_sales}")
print(f"Variance: {variance}")
```

#### Step 7: Drill-Through Analysis: Sales by Product Subcategory or Customer Segment

```python
# Sales by product subcategory
subcategory_sales = superstore_data.groupby('Product Sub-Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
subcategory_sales.plot(kind='bar', color='cornflowerblue')
plt.title('Sales by Product Sub-Category')
plt.xlabel('Product Sub-Category')
plt.ylabel('Total Sales')
plt.show()
```

#### Step 8: Sales by Product Category

###### total sales by the main product categories

#### Profitability by Product Sub-Category

###### the most profitable subcategories to understand where the business earns the highest profit.

```python
# Calculate total profit by subcategory
subcategory_profit = superstore_data.groupby('Product Sub-Category')['Profit'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 8))
subcategory_profit.plot(kind='bar', color='salmon')
plt.title('Profit by Product Sub-Category')
plt.xlabel('Product Sub-Category')
plt.ylabel('Total Profit')
plt.show()
```

#### Sales and Profit Trend Over Time

###### the trend of sales and profit over time to spot any seasonality or growth patterns.

```python
# Group by Order Month for time series analysis
monthly_sales_profit = superstore_data.groupby(superstore_data['Order Date'].dt.to_period('M')).agg({'Sales': 'sum', 'Profit': 'sum'})

# Plot sales and profit trend
monthly_sales_profit.plot(figsize=(14, 8), marker='o')
plt.title('Sales and Profit Trend Over Time')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend(['Sales', 'Profit'])
plt.show()
```

#### Customer Segment Analysis

###### customer segments to see which segments generate the most sales and profit.

```python
# Group by Customer Segment
segment_analysis = superstore_data.groupby('Customer Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values(by='Sales', ascending=False)

# Plot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
segment_analysis['Sales'].plot(kind='bar', ax=ax[0], color='lightblue')
segment_analysis['Profit'].plot(kind='bar', ax=ax[1], color='lightcoral')

ax[0].set_title('Sales by Customer Segment')
ax[1].set_title('Profit by Customer Segment')
plt.show()
```

#### Sales Forecasting

```python
# Import libraries
from statsmodels.tsa.arima.model import ARIMA
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Prepare monthly sales data
monthly_sales = superstore_data.resample('M', on='Order Date')['Sales'].sum()

# Fit an ARIMA model (you may need to tune parameters)
model = ARIMA(monthly_sales, order=(1, 1, 1))
model_fit = model.fit()

# Forecast sales for the next 12 months
forecast = model_fit.forecast(steps=12)
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales, label="Historical Sales")
plt.plot(forecast, label="Forecasted Sales", color="orange")
plt.legend()
plt.title('Sales Forecast for Next 12 Months')
plt.show()
```

#### Customer Segmentation for Targeted Marketing

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Aggregating customer data
customer_data = superstore_data.groupby('Customer ID').agg({
    'Sales': 'sum',
    'Order ID': 'nunique',
    'Discount': 'mean'
}).rename(columns={'Order ID': 'Frequency'})

# Scaling features
scaler = StandardScaler()
customer_data_scaled = scaler.fit_transform(customer_data)

# Applying KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(customer_data_scaled)

# Visualize clusters
plt.figure(figsize=(8, 6))
plt.scatter(customer_data['Sales'], customer_data['Frequency'], c=customer_data['Cluster'], cmap='viridis')
plt.xlabel('Total Sales')
plt.ylabel('Order Frequency')
plt.title('Customer Segmentation')
plt.show()
```

#### Discount Impact on Sales and Profit Prediction

```python
# Display column names to confirm if 'Quantity' or a similar column is available
print(superstore_data.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define features and target variable with the correct column names
X = superstore_data[['Discount', 'Quantity ordered new']]
y = superstore_data['Profit']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fit model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict and evaluate
predictions = model.predict(X_test)
print("Mean Squared Error:", mean_squared_error(y_test, predictions))
print("R-squared Score:", r2_score(y_test, predictions))
```

#### Predicting High-Risk Orders for Profit Loss Prevention

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Define high-risk (loss) orders
superstore_data['HighRisk'] = (superstore_data['Profit'] < 0).astype(int)
X = superstore_data[['Discount', 'Sales', 'Quantity ordered new']]
y = superstore_data['HighRisk']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fit model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```
